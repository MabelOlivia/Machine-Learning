```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = ['Monday', "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
y = [50, 51, 52, 48, 47, 49, 46]
```

```python
plt.plot(x,y, 'rD-.')
```

```python
plt.plot(x,y, color = "red", marker = "D", markersize = 8, linestyle = "dashdot")
```

```python
plt.plot(x,y, color = "Violet", alpha = 0.5)
```
