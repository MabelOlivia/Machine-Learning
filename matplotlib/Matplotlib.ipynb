```python
plt.plot(x,y, "ro--", linewidth = 3 , markersize = 10)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph in 2D")
```

```python
plt.subplot(2,2,2)
plt.plot(x,y, "r*--")
plt.subplot(2,2,1)
plt.plot(x,y, "b*--")
plt.subplot(2,2,3)
plt.plot(x,y, "g*")
```

```python
x = np.arange(1, 11)
y = 3 * x + 5
plt.plot(x,y)
```

```python
# Sine wave form
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title("Sine waveform")
plt.plot(x,y)
```

```python
# subplot

x = np.arange(0, np.pi*5, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin, "g*--")
plt.title("Sine wave")

plt.subplot(2,1,2)
plt.plot(x,y_cos, "r*--")
plt.title("cosine wave")
```

```python
## Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

## Pie Plot

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
