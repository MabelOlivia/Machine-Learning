```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = ['Monday', "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
y = [50, 51, 52, 48, 47, 49, 46]
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, color = "Pink", linewidth=4, linestyle = "dashdot")
```
