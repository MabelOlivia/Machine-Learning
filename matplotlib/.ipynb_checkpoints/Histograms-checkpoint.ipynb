```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins =[80,100,125,150] , rwidth = 0.95, color = "magenta")
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel("sugar level range")
plt.ylabel("Total number of patients")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_men,blood_sugar_women], bins = [80,100,125,150], rwidth = 0.95, color = ["magenta", "green"],
        label = ["men", "women"])
plt.legend()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel("Total number of patients")
plt.ylabel("sugar level range")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_men,blood_sugar_women], bins = [80,100,125,150], rwidth = 0.95, color = ["magenta", "green"],
        label = ["men", "women"], orientation = "horizontal")
plt.legend()
```
