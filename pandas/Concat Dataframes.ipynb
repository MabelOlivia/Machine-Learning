```python
temperature_weather = pd.DataFrame({
    "city": ["Nairobi", "Mombasa", "Kisumu"],
    "temperature": [26, 30, 28]
})
temperature_weather
```

```python
wind_speed = pd.DataFrame({
    "city": ["Kisumu", "Nairobi", "Mombasa"],
    "windspeed": [8, 9, 10]
})
wind_speed
```

```python
df = pd.merge(temperature_weather, wind_speed, on = "city")
df
```

```python
temp_df = pd.DataFrame({
    "city": ["Nairobi", "Mombasa", "kakamega", "Nyeri", "Embu"],
    "temperature": [26, 30, 28, 24, 27]
})
temp_df
```

```python
wind_df = pd.DataFrame({
    "city": ["Nairobi", "Juja", "Mombasa"],
    "windspeed": [8, 9, 10]
})
wind_df
```

```python
df = pd.merge(temp_df, wind_df, on = "city")
df
```

```python
df = pd.merge(temp_df, wind_df, on = "city", how = "outer")
df
```

```python
df = pd.merge(temp_df, wind_df, on = "city", how = "right")
df
```

```python
df = pd.merge(temp_df, wind_df, on = "city", how = "left")
df
```

```python
df = pd.merge(temp_df, wind_df, on = "city", how = "outer", indicator = True)
df
```
