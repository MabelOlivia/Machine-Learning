```python
import pandas as pd 
df = pd.read_csv("titanic.csv")
df.head(4)
```

```python
df = df.drop(["PassengerId", "Name", "SibSp", "Pclass", "Parch", "Ticket", "Cabin", "Embarked" ], axis = "columns")
df.head()
```

```python
target = df[['Survived']]
target.head()
```

```python
features = df.drop(["Survived"], axis = "columns")
features.head()
```

```python
dummies = pd.get_dummies(df.Sex)
dummies.head(10)
```

```python
features = pd.concat([dummies, features], axis = "columns")
features.head()
```

```python
features = features.drop(["Sex"], axis = "columns")
features.head()
```

```python
features.columns[features.isna().any()]
```

```python
features.Age[:10]
```

```python
features.Age = features.Age.fillna(features.Age.median())
features.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train.values.ravel())
```

```python
y_test[:10]
```

```python
y_predicted = model.predict(X_test)
y_predicted[:10]
```

```python
model.score(X_test, y_test)
```
