# Iris Dataset

```python
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

## Correlation

```python
iris.corr()
```

```python
sns.heatmap(iris.corr())
```

```python
sns.jointplot("sepal_length", y= "sepal_width", data = iris)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue = "species")
```

```python
sns.countplot("species", data=iris)
```

```python
sns.jointplot("sepal_width", y= "petal_length", data = iris)
```
