```python
import pandas as pd
import numpy as np
df = pd.read_csv("titanic.csv")
df.head(10)
```

```python
target = df.Survived
```

```python
features = df.drop(["PassengerId", "Name", "SibSp", "Parch", "Ticket", "Cabin", "Embarked", "Survived"], axis = "columns")
features.head(10)
```

```python
feature = features.fillna(method = 'ffill') 
feature.head(10)
```

### An alternative

```python
feature.Age = feature.Age.fillna(feature.Age.mean())
```

```python
feature.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
feature["Sex_n"] = le.fit_transform(feature["Sex"])
```

```python
feature = feature.drop(["Sex"], axis = "columns")
feature
```

### ALTERNATIVE

```python
features.Sex = features.Sex.map({'male': 1, 'female': 2})
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size = 0.3)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```
